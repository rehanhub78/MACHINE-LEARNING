```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
X = np.random.rand(50,1)
Y=1.3*(X**8)-\
  1.4*(X**7)-\
  0.8*(X**6)+\
  0.6*(X**5)-\
  0.3*(X**4) + 0.5*(X**3) + 1.2*(X**2) - 0.6*X + 0.1*np.random.rand(50,1)
print(X)
```

```python
fig = plt.figure()
plt.scatter(X,Y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures

for i in range(1,8):
    poly = PolynomialFeatures(i)
    x_poly=poly.fit_transform(X)
    scaler=StandardScaler() 
    scaler.fit(x_poly)
    x_poly_stnd=scaler.transform(x_poly)
    model=LinearRegression()
    model.fit(x_poly_stnd,Y)
    output=model.predict(x_poly_stnd)
    fig=plt.figure()
    plt.scatter(X,output,label="predict")
    plt.scatter(X,Y,label="given")
    plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures

for i in range(1,20):
    poly = PolynomialFeatures(i)
    x_poly=poly.fit_transform(X)
    scaler=StandardScaler() 
    scaler.fit(x_poly)
    x_poly_stnd=scaler.transform(x_poly)
    model=LinearRegression()
    model.fit(x_poly_stnd,Y)
    output=model.predict(x_poly_stnd)
    fig=plt.figure()
    plt.scatter(X,output,label="predict")
    plt.scatter(X,Y,label="given")
    plt.show()
    '''
    upper code overfit the data and it is not good for prediction.
    to avoid overfitting we can use regularization techniques like ridge regression and lasso regression.
    In bias-variance tradeoff we can see that as we increase the complexity of the model, the bias decreases and the variance increases.
        In the above code we can see that as we increase the degree of the polynomial, the bias decreases and the variance increases.
    '''
```
