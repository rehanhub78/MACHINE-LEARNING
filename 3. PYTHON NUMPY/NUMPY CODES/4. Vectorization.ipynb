```python
import numpy as np
import math
m = np.arange(1,21).reshape(4,5)
print(m)
```

```python
# vectorization is the process of applying a function to an entire array, rather than applying it to each element individually.
# - This can be done using the np.vectorize() function, which takes a function as an argument and returns a new function that can be 
#       applied to arrays.


vecfun = np.vectorize(math.sqrt) # it will apply the sqrt function to each element of the array m.
print(vecfun(m))
vecfun = np.vectorize(math.sin) # it will apply the sin function to each element of the array m.
print(vecfun(m))
```

```python
def myfunc(a, b):
    "Return a-b if a > b, and a+b otherwise."
    if a > b:
        return a - b
    else:
        return a + b
```

```python
m2 = np.array([1,2,3,4])
vfunc = np.vectorize(myfunc)
print(vfunc(m2,2)) # it will apply the myfunc function to each element of the array m2, with 2 as the second argument.
```
