```python
'''
STACKING: Stacking is the process of combining multiple arrays into a single array. 
   -This can be done using the np.stack() function, which takes a sequence of arrays and an axis as arguments and returns a 
       new array that is the result of stacking the input arrays along the specified axis.
       '''
import numpy as np
A = np.arange(1,5).reshape(2,2)
B = np.arange(11,19).reshape(2,4)
print(A)
print(B.T)
```

```python
np.hstack((A,B)) # it will stack the arrays A and B horizontally (column-wise). so the number of rows should be same in both the arrays.
```

```python
np.vstack((A,B.T)) # it will stack the arrays A and B vertically (row-wise). so the number of columns should be same in both the arrays.
```
